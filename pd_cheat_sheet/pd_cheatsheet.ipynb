```python
#import pandas and vege_datasets libraries
import pandas as pd
from vega_datasets import data
```

```python
#list the datasets available
data.list_datasets()
```

```python
#store and view data
df = data.airports()
df
```

```python
#View first row of the data
df.head()
```

```python
#view the last rows of the data 
df.tail()
```

```python
#return data types of each column
df.dtypes
```

```python
#list all column names
df.columns
```

```python
#get a summer of numerical columns
df.describe()
```

```python
#Renaming the index column
df.index.name = 'id'
df
```

```python
#renaming specific columns
df.rename(columns={'latitude': 'lat'}, inplace=True)
df.head()
```

```python
#Rename multiple columns at a go
df.rename(columns= {'lat': 'latitude','longitude':'lon'}, inplace = True)
df
```

```python
#return column name lon to latitude
df.rename(columns={'lon': 'longitude'}, inplace=True)
```

```python
#sort values
df.sort_values('city', ascending = False)
```

```python
df.sort_values('country', ascending = True)
```

```python
#groupby and count the number in each group
df.groupby('city').count()
```

```python
#groupby 2 columns and count 
df.groupby(['country','city']).count()
```

```python
#groupby city and return the mean of th elatitude column
df.groupby('city')['latitude'].mean()
```

# Querry data

```python
# filter df to return rows with 'OH' in state column
ohio = df['state'] == 'OH'
```

```python
# return boolean values of the filter
ohio
```

```python
# convert boolean values to df
df[ohio]
```

```python
df_filtered = df[(df['iata'] == '10G') & (df['city'] == 'Millersburg')]
df_filtered
```

```python
df_filtered = df[(df['iata'] == '10G') | (df['city'] == 'Bryan')]
df_filtered
```

```python
df_filtered = df[(df['iata'] != '10G') & (df['city'] != 'Livingston')]
df_filtered
```
